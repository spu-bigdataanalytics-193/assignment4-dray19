```
from operator import add
```

```
rdd = sc.textFile('/FileStore/tables/*.bz2',minPartitions= 4)
rdd.take(4)
```

```
rdd1 = rdd.map(lambda x: x.split(',')[8]).filter(lambda x : 'UniqueCarrier' not in x)
rdd1.take(10)
```

```
rdd2 = rdd1.map(lambda x:(x,1))
rdd2.take(10)
```

```
rdd4 = rdd2.reduceByKey(add).map(lambda x: (x[1], x[0])).sortByKey(False)
rdd4.take(30)
```
